##  Hled√°n√≠ minima za vyu≈æit√≠ algoritmu z knihovny scipy.optimize

```python

true_min1=1.0
print("Opravdov√© minimum f = 1.0")
for n in [2,3,4,5,6,7,8,25]:
    h2d = Hilbert2D(10)
    t_min, h_min, f_min, nfev = h2d.find_minimum_mapped(n,0,1,0,1,0, true_min1, ftol=1e-6)
    print(f"precision={n:3d}  =>  t = {t_min:.6f},  h(t) = ({h_min[0]:.4f}, {h_min[1]:.4f}),  f = {f_min:.6f}")
```

```python
true_min2=-1.9133
print("Opravdov√© minimum f = -1.9133")

for n in [2,3,4,5,6,7,8,10,11,12,13]:
    h2d = Hilbert2D(10)
    t_min, h_min, f_min, nfev = h2d.find_minimum_mapped(n,0,1,0,1,1, true_min2, ftol=1e-6)
    print(f"precision={n:3d}  =>  t = {t_min:.6f},  h(t) = ({h_min[0]:.4f}, {h_min[1]:.4f}),  f = {f_min:.6f}")
```

```python
true_min3=0.0
print("Opravdov√© minimum f = 0.0")
for n in [2,3,4,5,6,7,8]:
    h2d = Hilbert2D(10)
    t_min, h_min, f_min, nfev = h2d.find_minimum_mapped(n,0,1,0,1,2, true_min3, ftol=1e-6)
    print(f"precision={n:3d}  =>  t = {t_min:.6f},  h(t) = ({h_min[0]:.4f}, {h_min[1]:.4f}),  f = {f_min:.6f}")
```
