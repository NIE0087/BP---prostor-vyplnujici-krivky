## Test mezivýpočtů pro výpočet bodu křivky 

```python

Q = hilbert.dec_to_quarter(0.2)
print("Quarter Q:", Q)

e0, dj = hilbert.ej_and_dj_counter(Q)
result = hilbert.calculate_point(e0, dj, Q)
print("Calculated point:", result)

```

```python

h = [2, 0, 3]
e0, dj = hilbert.ej_and_dj_counter(h)
print("e0:", e0)
print("dj:", dj)

result = hilbert.calculate_point(e0, dj, h)
print("Point from [2,0,3]:", result)
```

## Pro Braninovu funkci

$$f_1(u,v)=a\,(v-bu^2+cu-r)^2+s(1-t)\cos(u)+s$$

kde

$$a=1,\quad b=\frac{5.1}{4\pi^2},\quad c=\frac{5}{\pi},\quad r=6,\quad s=10,\quad t=\frac{1}{8\pi}$$

Mapování z jednotkového čtverce:

$$u=-5+10x,\qquad v=-5+10y$$

Globální minimum: $f_{\min}\approx 0.3978873577$.
